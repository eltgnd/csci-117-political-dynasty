```python
import json

def get_unique_pairs(filepath):
    unique_pairs = set()
    
    # Load the GeoJSON file
    with open(filepath, 'r', encoding='utf-8') as file:
        data = json.load(file)
        
    # Loop through each feature to extract the properties
    for feature in data.get('features', []):
        properties = feature.get('properties', {})
        
        municipality = properties.get('NAME_2')
        region = properties.get('NAME_1')
        
        # Add to our set (which automatically handles duplicates)
        if municipality and region:
            unique_pairs.add((municipality, region))
            
    return sorted(list(unique_pairs))

# --- HOW TO USE ---
# 1. Replace 'philippines.geojson' with the actual path to your file
file_path = 'data/gadm41_PHL_2.json' 

pairs = get_unique_pairs(file_path)

# Print the list in a cleanly formatted way so you can easily copy/paste it
print("[")
for pair in pairs:
    print(f"    {pair},")
print("]")
```
